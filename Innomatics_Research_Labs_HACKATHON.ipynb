```python
import pandas as pd
import json
import re
print("--- Loading Data ---")
df_orders = pd.read_csv('orders.csv')
with open('users.json', 'r') as f:
    users_data = json.load(f)
df_users = pd.DataFrame(users_data)
with open('restaurants.sql', 'r') as f:
    sql_content = f.read()
pattern = r"\((\d+),\s*'([^']*)',\s*'([^']*)',\s*([\d\.]+)\)"
matches = re.findall(pattern, sql_content)
df_restaurants = pd.DataFrame(matches, columns=['restaurant_id', 'restaurant_name_sql', 'cuisine', 'rating'])
df_restaurants['restaurant_id'] = df_restaurants['restaurant_id'].astype(int)
df_restaurants['rating'] = df_restaurants['rating'].astype(float)
df_final = df_orders.merge(df_users, on='user_id', how='left')
df_final = df_final.merge(df_restaurants, on='restaurant_id', how='left')
df_final.to_csv('final_food_delivery_dataset.csv', index=False)
print("Success: final_food_delivery_dataset.csv created.\n")

print("--- Numerical Answers ---")
gold_orders = df_final[df_final['membership'] == 'Gold']
print(f"1. Total Gold orders: {len(gold_orders)}")

hyd_rev = df_final[df_final['city'] == 'Hyderabad']['total_amount'].sum()
print(f"2. Hyderabad Revenue: {round(hyd_rev)}")
print(f"3. Distinct users: {df_final['user_id'].nunique()}")
avg_gold = gold_orders['total_amount'].mean()
print(f"4. Avg Order Value (Gold): {round(avg_gold, 2)}")
print(f"5. Orders with rating >= 4.5: {len(df_final[df_final['rating'] >= 4.5])}")
top_gold_city = gold_orders.groupby('city')['total_amount'].sum().idxmax()
top_gold_city_orders = len(gold_orders[gold_orders['city'] == top_gold_city])
print(f"6. Orders in top Gold city ({top_gold_city}): {top_gold_city_orders}\n")
print("--- MCQ Selection Answers ---")

print(f"Cuisine with highest Avg Order Value: {df_final.groupby('cuisine')['total_amount'].mean().idxmax()}")

user_spend = df_final.groupby('user_id')['total_amount'].sum()
print(f"Users with total spend > 1000: {(user_spend > 1000).sum()} (Select: > 2000)")
pct_gold = (len(gold_orders) / len(df_final)) * 100
print(f"Percentage of Gold orders: {round(pct_gold)}%")
df_final['order_date'] = pd.to_datetime(df_final['order_date'], dayfirst=True)
df_final['quarter'] = df_final['order_date'].dt.quarter
print(f"Highest Revenue Quarter: Q{df_final.groupby('quarter')['total_amount'].sum().idxmax()}")
```
